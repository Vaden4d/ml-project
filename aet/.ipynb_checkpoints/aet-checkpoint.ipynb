```python
def accuracy(y_pred, y_true):
  
    y_true = y_true.argmax(dim=1)
    y_pred = y_pred.argmax(dim=1)
  
    binary = y_true == y_pred
    return binary.sum().double() / binary.size()[0]
```

> <font size=4> Below the process of defining the pretrained model: deep convolutional model with a lot of blocks. This is code from original authors GitHub. Classes PretrainedModel and Classifier are our wrappers for using the representations. Additionaly were defined functions which provide an opportunity to refer to the model's intermediate layers by name. </font>

```python
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F

class BasicBlock(nn.Module):
    def __init__(self, in_planes, out_planes, kernel_size):
        super(BasicBlock, self).__init__()
        padding = (kernel_size-1) // 2
        self.layers = nn.Sequential()
        self.layers.add_module('Conv', nn.Conv2d(in_planes, out_planes, \
            kernel_size=kernel_size, stride=1, padding=padding, bias=False))
        self.layers.add_module('BatchNorm', nn.BatchNorm2d(out_planes))
        self.layers.add_module('ReLU',      nn.ReLU(inplace=True))

    def forward(self, x):
        return self.layers(x)

        feat = F.avg_pool2d(feat, feat.size(3)).view(-1, self.nChannels)

class GlobalAveragePooling(nn.Module):
    def __init__(self):
        super(GlobalAveragePooling, self).__init__()

    def forward(self, feat):
        num_channels = feat.size(1)
        return F.avg_pool2d(feat, (feat.size(2), feat.size(3))).view(-1, num_channels)
      
class AveragePooling(nn.Module):
    def __init__(self):
        super(AveragePooling, self).__init__()
                
    def forward(self, feat, kernel_size):
        return F.avg_pool2d(feat, kernel_size)

# entire model class    
class NetworkInNetwork(nn.Module):
    def __init__(self, _num_inchannels=3, _num_stages=3, _use_avg_on_conv3=True):
        super(NetworkInNetwork, self).__init__()

        num_inchannels = _num_inchannels
        num_stages = _num_stages
        use_avg_on_conv3 = _use_avg_on_conv3


        assert(num_stages >= 3)
        nChannels  = 192
        nChannels2 = 160
        nChannels3 = 96

        blocks = [nn.Sequential() for i in range(num_stages)]
        # 1st block
        blocks[0].add_module('Block1_ConvB1', BasicBlock(num_inchannels, nChannels, 5))
        blocks[0].add_module('Block1_ConvB2', BasicBlock(nChannels,  nChannels2, 1))
        blocks[0].add_module('Block1_ConvB3', BasicBlock(nChannels2, nChannels3, 1))
        blocks[0].add_module('Block1_MaxPool', nn.MaxPool2d(kernel_size=3,stride=2,padding=1))

        # 2nd block
        blocks[1].add_module('Block2_ConvB1',  BasicBlock(nChannels3, nChannels, 5))
        blocks[1].add_module('Block2_ConvB2',  BasicBlock(nChannels,  nChannels, 1))
        blocks[1].add_module('Block2_ConvB3',  BasicBlock(nChannels,  nChannels, 1))
        blocks[1].add_module('Block2_AvgPool', nn.AvgPool2d(kernel_size=3,stride=2,padding=1))

        # 3rd block
        blocks[2].add_module('Block3_ConvB1',  BasicBlock(nChannels, nChannels, 3))
        blocks[2].add_module('Block3_ConvB2',  BasicBlock(nChannels, nChannels, 1))
        blocks[2].add_module('Block3_ConvB3',  BasicBlock(nChannels, nChannels, 1))

        if num_stages > 3 and use_avg_on_conv3:
            blocks[2].add_module('Block3_AvgPool', nn.AvgPool2d(kernel_size=3,stride=2,padding=1))
        for s in range(3, num_stages):
            blocks[s].add_module('Block'+str(s+1)+'_ConvB1',  BasicBlock(nChannels, nChannels, 3))
            blocks[s].add_module('Block'+str(s+1)+'_ConvB2',  BasicBlock(nChannels, nChannels, 1))
            blocks[s].add_module('Block'+str(s+1)+'_ConvB3',  BasicBlock(nChannels, nChannels, 1))

        # global average pooling and classifier
        blocks.append(nn.Sequential())
        blocks[-1].add_module('GlobalAveragePooling',  GlobalAveragePooling())

        self._feature_blocks = nn.ModuleList(blocks)
        self.all_feat_names = ['conv'+str(s+1) for s in range(num_stages)] + ['classifier',]
        assert(len(self.all_feat_names) == len(self._feature_blocks))

        self.weight_initialization()

    def _parse_out_keys_arg(self, out_feat_keys):

        # By default return the features of the last layer / module
        out_feat_keys = [self.all_feat_names[-1],] if out_feat_keys is None else out_feat_keys

        if len(out_feat_keys) == 0:
            raise ValueError('Empty list of output feature keys.')
        for f, key in enumerate(out_feat_keys):
            if key not in self.all_feat_names:
                raise ValueError('Feature with name {0} does not exist. Existing features: {1}.'.format(key, self.all_feat_names))
            elif key in out_feat_keys[:f]:
                raise ValueError('Duplicate output feature key: {0}.'.format(key))

        # Find the highest output feature in `out_feat_keys
        max_out_feat = max([self.all_feat_names.index(key) for key in out_feat_keys])

        return out_feat_keys, max_out_feat

    def forward(self, x, out_feat_keys=None):
        """Forward an image `x` through the network and return the asked output features.

        Args:
          x: input image.
          out_feat_keys: a list/tuple with the feature names of the features
                that the function should return. By default the last feature of
                the network is returned.

        Return:
            out_feats: If multiple output features were asked then `out_feats`
                is a list with the asked output features placed in the same
                order as in `out_feat_keys`. If a single output feature was
                asked then `out_feats` is that output feature (and not a list)."""
        out_feat_keys, max_out_feat = self._parse_out_keys_arg(out_feat_keys)
        out_feats = [None] * len(out_feat_keys)

        feat = x
        for f in range(max_out_feat+1):
            feat = self._feature_blocks[f](feat)
            key = self.all_feat_names[f]
            if key in out_feat_keys:
                out_feats[out_feat_keys.index(key)] = feat
        out_feats = out_feats[0] if len(out_feats) == 1 else out_feats
        return out_feats

    def weight_initialization(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                if m.weight.requires_grad:
                    n = m.kernel_size[0] * m.kernel_size[1] * m.out_channels
                    m.weight.data.normal_(0, np.sqrt(2. / n))
            elif isinstance(m, nn.BatchNorm2d):
                if m.weight.requires_grad:
                    m.weight.data.fill_(1)
                if m.bias.requires_grad:
                    m.bias.data.zero_()
            elif isinstance(m, nn.Linear):
                if m.bias.requires_grad:
                    m.bias.data.zero_()

# model that connected transformation part encoder and ordinal encoder                    
class Regressor(nn.Module):
    def __init__(self, _num_stages=3, _use_avg_on_conv3=True, 
                 indim=384, num_classes=6):
        super(Regressor, self).__init__()
        self.nin = NetworkInNetwork(_num_stages=_num_stages, 
                                    _use_avg_on_conv3=_use_avg_on_conv3)
        self.fc = nn.Linear(indim, num_classes)
        for m in self.modules():
            if isinstance(m, nn.Linear):
                m.bias.data.zero_()

    def forward(self, x1, x2, out_feat_keys=None):
        x1 = self.nin(x1, out_feat_keys)
        x2 = self.nin(x2, out_feat_keys)
        if out_feat_keys==None:
            x = torch.cat((x1,x2), dim=1)
            return x1, x2, torch.tanh(self.fc(x))
        else:
            return x1, x2
          
class Flatten(nn.Module):
    def __init__(self):
        super(Flatten, self).__init__()

    def forward(self, feat):
        return feat.view(feat.size(0), -1)

# class-wrapper for pretrained model
class PretrainedModel(nn.Module):
    def __init__(self, out_feat_key):
      
        super(PretrainedModel, self).__init__()
        self.output_layer = [out_feat_key]
      
        weights_dict = torch.load('output/net_epoch_1499.pth')

        modified_weights = OrderedDict()
        for x, y in zip(weights_dict.keys(), weights_dict.values()):
            modified_weights[x[7:]] = y

        self.pretrained_model = Regressor(_num_stages=4, 
                                   _use_avg_on_conv3=True)
        self.pretrained_model.load_state_dict(modified_weights) 
        self.pretrained_model = self.pretrained_model.nin
      
        self.pooling = AveragePooling()
      
        self.flatten = Flatten()
      
    def forward(self, x):
      
        output = self.pretrained_model(x, self.output_layer)
        output = self.flatten(output)
      
        return output

# classificator on the top of pretrained model
class Classificator(nn.Module):
    def __init__(self, pretrained_model, n_classes=10):
      
        super(Classificator, self).__init__()
        self.pretrained_model = pretrained_model
      
        for param in self.pretrained_model.parameters():
            param.requires_grad = False
          
        self.conv = nn.Conv2d(192, 81, (3, 3))
        self.flatten = Flatten()
        self.linear_1 = nn.Linear(324, 200)
        self.bn_1 = nn.BatchNorm1d(200)
        self.linear_2 = nn.Linear(200, 50)
        self.bn_2 = nn.BatchNorm1d(50)
        self.linear_3 = nn.Linear(50, 10)
      
        self.relu = nn.ReLU()
           
    def forward(self, x):
      
        x = self.pretrained_model(x)
      
        x = self.conv(x)
        x = self.flatten(x)
      
        x = self.linear_1(x)
        x = self.bn_1(x)
        x = self.relu(x)
      
        x = self.linear_2(x)
        x = self.bn_2(x)
        x = self.relu(x)
      
        x = self.linear_3(x)
      
        return x
```

> <font size=4> The code below is aggregated train/test updated for the model in the future training process. </font>

```python
class Trainer:
    def __init__(self, model, optimizer, criterion, metric, clip_norm, device):

        self.device = device
        
        self.model = model.cuda() if self.device.type == 'cuda' else model
        self.optimizer = optimizer
        self.criterion = criterion
        self.clip_norm = clip_norm
        self.metric = metric
        self.num_updates = 0
        
        self.activation = nn.Softmax()

    def train_step(self, batch):

        self.model.train()
        self.optimizer.zero_grad()
        self.num_updates += 1
        
        features, responses_real = batch
        responses_pred = self.activation(self.model(features))
        
        loss = self.criterion(responses_pred, responses_real)
        self.backward(loss)
        
        quality = self.metric(responses_pred, responses_real)
        return loss, quality

    def test_step(self, batch):

        self.model.eval()
        features, responses_real = batch
        responses_pred = self.activation(self.model(features))

        loss = self.criterion(responses_pred, responses_real)
        quality = self.metric(responses_pred, responses_real)
        return loss, quality

    def backward(self, loss):

        loss.backward()
        torch.nn.utils.clip_grad_norm_(self.model.parameters(), self.clip_norm)
        self.optimizer.step()
```

> <font size=4> As we download weights to the initialized model, we can exctract the features from images. The model was trained using train part of CIFAR, hence we will visualize the similarity between images on the test part of data. </font>

```python
%%time

vectors = []

# for every picture save the vector representation into the vectors list and then save as toch.Tensor (for convention)

for image, label in testset:
  
    vector = obj(image.expand(1, 3, 32, 32).cuda()).cpu().detach().numpy()
    vectors.append(vector)
    
vectors = np.squeeze(np.array(vectors))

torch.save(torch.Tensor(vectors), '../../vectors/vectors_blocknet_conv3.pth')
```

> <font size=4> We validated the model on the test set, so it will be used in validation for visualization purposes (to avoid overfitting and wrong conclusions). As we have compressed vector equivalent for every picture in the test set, we can construct the kNN algorithm on the top of this vectors. We will use the KDTree algorithm - the optimized version of the vanilla kNN (thanks to the Scikit-learn). </font>

```python
%%time

n_neighbors = 5

knn = KDTree(vectors)
# compute the matrix with index of an image as a row and columns - indexes of the closest images by the norm
# of course, norm of difference between vector representations
distances, indexes = knn.query(vectors, n_neighbors)

np.save('indexes-conv3.npy', indexes)
```

```python
indexes = np.load('indexes-conv3.npy')
```

```python
def plot_examples_knn(data, indexes, nrows=4, nexamples=3, save=True):
    '''Plot the closest data points 
    of the randomly chosen data points.
    
    Parameters
    ----------
    data : PyTorch DataLoader, or arbitary indexed type with Tensors
        i.e data points that used in similarity search task
    indexes : np.ndarray
        The indexes of the closest objects.Built by kNN object.
    nrows, nexamples : int
        The number of examples and their neighbors
    '''
    data_size = len(data)
        
    # randomly choose the pictures to visualize
    chosen_indexes = np.random.choice(data_size, nrows, replace=False)
    
    subset = indexes[chosen_indexes].astype(int)
    
    # plot the clusters and its elements in the grid
    fig, axes = plt.subplots(nrows, nexamples+1, figsize=(16, 16), dpi=100)
    # plot examples
    for i in range(nrows):   
        for j in range(nexamples+1):
            
            #pytorch tensors preparation
            picture = data[subset[i][j]][0].squeeze().permute(1, 2, 0).numpy()
            label = classes[data[subset[i][j]][1]]
            
            # plot pictures
            axes[i, j].axis('off')
            axes[i, j].imshow(picture)
            
            if j > 0:
                axes[i, j].set_title('Neighbor {y} for {x}, \n class - {z}'.format(x=subset[i][0], y=subset[i][j], z=label))
            else:
                axes[i, j].set_title('Picture {y}, \n class - {x}'.format(x=label, y=subset[i][0]))
                
    if save:
        fig.savefig('similarities-aet.png', dpi=150)
```

> <font size=4> Will train the own classifier on the top of extracted features to prove conclusion above. Before fitting images into the model, transform data as defined in transformations below. Optimizer - Adam. The metaparameters for training defined in cells. </font>

```python
## training classificator on the top
n_epochs = 100
batch_size = 32
```

```python
transform_train = transforms.Compose([
        transforms.RandomCrop(32, padding=4),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010)),
    ])

    transform_test = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010)),
    ])

trainset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform_train)
train_loader = torch.utils.data.DataLoader(trainset, batch_size=batch_size,
                                          shuffle=True, num_workers=2)

testset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                       download=True, transform=transform_test)
test_loader = torch.utils.data.DataLoader(testset, batch_size=batch_size,
                                         shuffle=True, num_workers=2)

classes = ('plane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

```python
model = Classificator(obj)
```

```python
lr = 1e-2

optimizer = Adam(model.parameters(), lr)
scheduler = ReduceLROnPlateau(optimizer, 'min', patience=10, eps=1e4)
clip_norm = 1.0
criterion = BCEWithLogitsLoss()
metric = accuracy

trainer = Trainer(model, optimizer, criterion, metric, clip_norm, device)

n_classes = 10
one_hot_classes = torch.eye(n_classes)

test_iter = iter(test_loader)

j = 0
for i, epoch in enumerate(range(n_epochs)):
        
    for train_batch in train_loader:
        
        j += 1
        train_images, train_labels = train_batch
        try:
            test_images, test_labels = next(test_iter)
        except StopIteration:
            test_iter = iter(test_loader)
              test_images, test_labels = next(test_iter)
          
            train_labels = one_hot_classes[train_labels]
            test_labels = one_hot_classes[test_labels]
        
            train_loss, train_acc = trainer.train_step((train_images.cuda(), train_labels.cuda()))
            test_loss, test_acc = trainer.test_step((test_images.cuda(), test_labels.cuda()))
        
            if j % 1000 == 0:
                print('Iteration - {x}, epoch - {y}'.format(x=trainer.num_updates, y=i+1))
                print('Train loss - {x:.5f}, test loss - {y:.5f}'.format(x=train_loss, y=test_loss))
                print('Train accuracy - {x:.5f}, test accuracy - {y:.5f}'.format(x=train_acc, y=test_acc))
                print('LR - {}'.format(lr))

    scheduler.step(test_loss)
    lr = [group['lr'] for group in optimizer.param_groups][0]
```

```python
from AlexNet import Regressor

class Flatten(nn.Module):
    def __init__(self):
        super(Flatten, self).__init__()

    def forward(self, feat):
        return feat.view(feat.size(0), -1)
          
class PretrainedModel(nn.Module):
    def __init__(self, out_feat_key):
      
        super(PretrainedModel, self).__init__()
        self.output_layer = [out_feat_key]
      
        weights_dict = torch.load('output/net_epoch_99.pth', map_location='cpu')

        modified_weights = OrderedDict()
        for x, y in zip(weights_dict.keys(), weights_dict.values()):
            modified_weights[x[7:]] = y

        self.pretrained_model = Regressor(6)
        self.pretrained_model.load_state_dict(modified_weights) 
        self.pretrained_model = self.pretrained_model.Alexnet
            
        self.flatten = Flatten()
      
    def forward(self, x):
      
        output = self.pretrained_model(x, self.output_layer)      
        output = self.flatten(output)
      
        return output
```
