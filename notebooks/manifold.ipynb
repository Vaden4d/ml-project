```python
import numpy as np
import torch
import umap

import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline

import torchvision
import torchvision.transforms as transforms

from tqdm import tqdm_notebook as tqdm
```

### Setting up

> The only one thing you need to specify is the path to vectors obtained from unsupervised feature extractor. <br>
  Also, keep in mind that if you've already ran train script, then you would have CIFAR-10 at **../data/**, otherwise it will be installed in the same folder.

```python
vectors_path = '../vectors/vectors.pth'
```

### Load images and labels

```python
# define loader
transform = transform = transforms.Compose([transforms.Resize((224, 224)), transforms.ToTensor()])
dataset = torchvision.datasets.CIFAR10(root='../data', train=False, download=True, transform=transform)
loader = torch.utils.data.DataLoader(dataset, batch_size=1000, shuffle=False)

# load data
images = []
labels = []
for batch in tqdm(loader):
    images_, labels_ = batch
    
    images_ = images_.data.numpy()
    images_ = np.swapaxes(images_, 1, 3)
    images_ = np.swapaxes(images_, 1, 2)
    
    labels_ = labels_.data.numpy()
    
    images.extend(images_)
    labels.extend(labels_)

# convert label to it's name
label_to_name = {
    '0': 'airplane',
    '1': 'automobile',
    '2': 'bird',
    '3': 'cat',
    '4': 'deer',
    '5': 'dog',
    '6': 'frog',
    '7': 'horse',
    '8': 'ship',
    '9': 'truck' 
}  

labels = list(map(lambda x: label_to_name[str(x)], labels))
```

### Load unsupervised vectors

```python
vectors = torch.load(vectors_path)
vectors = vectors.data.numpy()
print('Number of vectors: {}'.format(vectors.shape[0]))
print('Vector size: {}'.format(vectors.shape[1]))
```

### Get embeddings

```python
# this might take some time...
random_state = np.random.RandomState(42)
embeddings = umap.UMAP(n_components=2, random_state=random_state).fit_transform(vectors)
```

### Visualize manifold

```python
fig, ax = plt.subplots(figsize=(20, 20))
_ = sb.scatterplot(x=embeddings[:, 0], y=embeddings[:, 1], hue=labels, ax=ax)
_ = ax.set_title('Representations manifold')
_ = ax.axis('off')
```

> <font size=3>
    Here we can observe the problem that we saw in similiarity search, but in large scale. There are no easy distinguishable clusters for each class. But still each representations has a couple of same-class neighbours, so similiarity search is still possible.
